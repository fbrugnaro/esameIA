```
import pandas as pd
from sklearn.model_selection import train_test_split
```

Read the Appointment csv

```
data = pd.read_csv("appointment_data.csv")
data
```

Prepare the dataset for the splitting

```
X, y = data.drop(['No-show'],axis=1), data['No-show'] 
X
```

```
y
```

Split the dataset for training and testing

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
```

```
X_train
```

```
X_test
```

```
y_train
```

```
y_test
```

Now that we have splitted the dataset we want to save the train and test set to csv.
Since it has to be a single csv for each set we have to join X and y

```
train = y_train.to_frame().join(X_train)
train
```

```
test = y_test.to_frame().join(X_test)
test
```

Now that we have a single dataframe for train and test we can save it to csv, index is set to false to avoid pandas from adding its index to the csv.

```
train.to_csv("train.csv", index=False)
```

```
test.to_csv("test.csv", index=False)
```

We load the newly saved csv and see that they have the correct dimensions

```
datatrain = pd.read_csv("train.csv")
datatrain
```
